```python
from pathlib import Path, PurePath
import matplotlib.pyplot as plt
```

```python
def myplot(title, names, steps):
    results_dir = "/data/natsuki/results"
    for name in names:
        names[name]['fid'] = list()
    alias = dict()

    for name in names:
        for i in steps:
            filename = PurePath(results_dir, name, f"test_s{i}", "fid")
            try:
                with open(filename) as f:
                    fid = float(f.read())
                    names[name]['fid'].append(fid)
            except:
                names[name]['fid'].append(None)
    msteps = list(step/1000000 for step in steps)
    plt.rcParams["figure.dpi"] = 100
    plt.grid()
    plt.xticks(msteps[1::2])
    plt.xlabel("million images to train")
    plt.ylabel("FID")
    plt.title(title)
    for i, name in enumerate(names):
        plt.plot(
            msteps, names[name]['fid'],
            label=names[name].get('label', name),
        )
    plt.legend()
```

```python
myplot(
    "SPADE vs BLADE",
    {
     "migan_nomonochrome_single": {'label': 'SPADE'},
#     "v1blade_nomonochrome_single": {'label': 'BLADEv1'},
     "v2blade_nomonochrome_single": {'label': 'BLADE'},
    },
    range(100000, 2000000 + 1, 100000)
)
```

```python
myplot(
    "Improve Learning stability at high resolution",
    {
     "512_v2blade_nomonochrome_single": {'label': 'BLADE'},
     "512_clade_nomonochrome_single": {'label': 'CLADE'},
     "512_bnspade_nomonochrome_single": {'label': 'SPADE+BatchNorm'},
     "512_v2bladeGskip_nomonochrome_single": {'label': 'BLADE+SkipInputs'},
    },
    range(100000, 2000000 + 1, 100000)
)
```

```python
myplot(
    "AlacGAN vs Ours",
    {
     "512_v2bladeGskip_nomonochrome_single": {'label': 'BLADE+SkipInputs'},
     "512_alacgan_nomonochrome_single": {'label': 'AlacGAN'},
    },
    range(100000, 2000000 + 1, 100000)
)
```

```python
myplot(
    "Ongoing Experiments",
    {
     "512_v2bladeGskip_nomonochrome_single": {'label': 'BLADE+SkipInputs'},
     "512_jpumagan_nomonochrome_single": {'label': 'Pseudo Segmentation'},
     "512_modulatedv2bladeGskip_nomonochrome_single": {'label': 'Global Style'},
    },
    range(100000, 2000000 + 1, 100000)
)
```
